```python
import pandas as pd
df = pd.read_csv("C:\\Users\\durga\\OneDrive\\Desktop\\Data Analytics\\streetwear_sales_data - streetwear_sales_data.csv")
df
```

1. What is the total sales revenue for each product.

```python
for index, row in df.iterrows():
    print(row["Product Name"],"=","$",row["Amount($)"])
```

2. which item generated the highest total revenue?

```python
df.groupby("Product Name")["Amount($)"].sum().sort_values(ascending=False).head(1)
```

3. Which product category (e.g., Sneakers, Hoodies) or brand performed best in units sold ?

```python
top_product = df.groupby("Product Name")["Quantity"].sum().idxmax()
print(f"Most sold item is {top_product}")
```

4. Which is brand is getting most revenue ?

```python
top_brand = df.groupby("Brand")["Amount($)"].sum().idxmax()
print(f"The brand generating the most revenue is {top_brand}")
```

5. Which product type does people buy more ?

```python
Most_bought=df.groupby("Product Type")["Quantity"].sum().idxmax()
print(f"The most bought product type is {Most_bought}")
```

6. Which country has the best sales and for which product ?

```python
Best_Country=df.groupby("Country")["Quantity"].sum().idxmax()
top_product_type= df.groupby("Product Type")["Quantity"].sum().idxmax()
print(f"The country with highest sales is {Best_Country} and most sold product type is {top_product_type}")
```

7. What's the least bought item is UK ?

```python
Uk_data = df.loc[df["Country"] == "UK"]
Least_Bought=Uk_data.groupby("Product Name")["Quantity"].sum().idxmin()
print(f"The least bought item in UK is {Least_Bought}")
```

8. How do sales vary between "Limited Edition" and regular products.

```python
Limited_edi = df.loc[df["Category"] == "Limited Edition"]["Amount($)"].sum()
Regular_edi = df.loc[df["Category"] != "Limited Edition"]["Amount($)"].sum()
print(f"The sales for limited edition is {Limited_edi} and sales for the regular or non-limited edition is {Regular_edi}")
```

9. Which payment method is the most used ?

```python
most_used_payment = df["Payment Mode"].value_counts().idxmax()
print(f"The most used payment method is {most_used_payment}")
```
